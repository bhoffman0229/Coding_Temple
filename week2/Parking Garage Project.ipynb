# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python

#  make a parking garage class

    #TAKE TICKET
#   car goes in, 
#           ----    available parking spaces goes down 
#                            ----    amount of tickets goes down

   #PAY FOR PARKING
#   customer inputs the money they owe in, it gets stored in a variable
#   If variable is not empty..... 
#                             ----  displays a message that their ticket has been paid  
#                                                                                        ----   and they have 15 min to leave
#                                       ----- updates currentticket dict key "paid" to True
   #LEAVEGARAGE
#   if ticket is paid, display "Thank You, have a nice day"
#   if not paid, display the input prompt for time spent in garage
#       ----    parkingSpaces list increases by 1
#                    ---- ticket amount list increases by 1

class parkinggarage:

    def __init__(self, tickets, parkingSpaces):
        self.tickets = tickets
        self.parkingSpaces = parkingSpaces
        self.currentTicket = {}
    
    def takeTicket(self):
        if self.parkingSpaces > 0:
            self.parkingSpaces -= 1
            self.tickets -= 1
            self.currentTicket[f'{str(self.tickets)}'] = False
            print('Thank you, You are number ' + str(abs((self.tickets)-150)) + '!')
        else:
            print("Please come back later, we are full at the moment. Thank you.")

    """Parking at this garage is free for the first hour, 
           5$ an hour afterwords. Payment by Card only         """
    def payForParking(self, number):
        new_number = str(-abs(number)+150)
        if self.currentTicket[new_number] == False:
            confirm = int(input('Please enter the number of hours spent in garage'))
            money_due = (confirm -1) *5
            if confirm < 2:
                print("All Set! You have 15 minutes to exit. Please come again")
            else:
                enterenter = input(f'Your balance is {money_due}$, please swipe your card and then type "enter" to continue')
                if enterenter.lower() == 'enter':
                    print("Payment Successful! You have 15 minutes to exit. Please come again :)")
                    self.currentTicket[new_number] = True
                else:
                    print('Please try payment again')
                
        else:
            print(f'Ticket number {number} is already paid :)')



    def leaveGarage(self, number1):
        new_number1 = str(-abs(number1)+150)
        if new_number1 is isinstance(self.currentTicket.keys):
            if self.currentTicket[new_number1] == True:
                self.tickets += 1
                self.parkingSpaces += 1
                del self.currentTicket[new_number1]
            
            else:
                self.payForParking(number1)
        else:
            print("Error, invalid ticket number")





Parking_Garage = parkinggarage(150,150)
Parking_Garage.takeTicket()
Parking_Garage.takeTicket()
Parking_Garage.takeTicket()
Parking_Garage.takeTicket()
Parking_Garage.takeTicket()
Parking_Garage.payForParking(1)
Parking_Garage.payForParking(5)
Parking_Garage.payForParking(4)
Parking_Garage.payForParking(2)
Parking_Garage.payForParking(3)
```
