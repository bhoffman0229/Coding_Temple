```python
df = pd.read_csv('/home/user/Documents/coding_temple/week6/day2/Homework6_2/taylor_utf-8.csv', encoding = 'UTF-8')
df.head()
```

```python
null_count = df.isnull().sum().sum()
print('Number of null values:', null_count)
```

```python
df.dropna(inplace=True)
null_count = df.isnull().sum().sum()
print('Number of null values:', null_count)
```

```python
df['Opening act(s)'] = df['Opening act(s)'].str.replace('\r\n',',')
df
```

```python
df = df.astype({'Attendance (tickets sold / available)':'string'})
df['Attendance (tickets sold / available)'].info
```

```python
df['Attendance (tickets sold / available)'] = df['Attendance (tickets sold / available)'].str.replace(' / ','_')
df['Attendance (tickets sold / available)'] = df['Attendance (tickets sold / available)'].str.replace(',','')
df['Attendance (tickets sold / available)']
```

```python
df[['tickets_sold','tickets_available']] = df['Attendance (tickets sold / available)'].str.split("_", expand = True)
del df['Attendance (tickets sold / available)']
df.head(10)
```

```python
df.columns = df.columns.str.replace(' ','_').str.lower().str.strip()
df
```
